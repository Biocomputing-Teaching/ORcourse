```python
import matplotlib.pyplot as plt
import numpy as np


def f(x):
  return np.sin(x)

def g(x, y):
  return x**4 - 8*x**3 + 24*x**2 - 32*x + 16

def h(x, y):
  return np.exp(-x**2 - y**2)

def plot_1d(func, x_range, title, xlabel, ylabel):
  x = np.linspace(x_range[0], x_range[1], 100)
  y = func(x)
  plt.figure()
  plt.plot(x, y)
  plt.title(title)
  plt.xlabel(xlabel)
  plt.ylabel(ylabel)
  plt.grid(True)
  plt.show()

def plot_2d(func, x_range, y_range, title, xlabel, ylabel, zlabel):
  x = np.linspace(x_range[0], x_range[1], 30)
  y = np.linspace(y_range[0], y_range[1], 30)
  X, Y = np.meshgrid(x, y)
  Z = func(X, Y)

  fig = plt.figure()
  ax = fig.add_subplot(111, projection='3d')
  ax.plot_surface(X, Y, Z, cmap='plasma')
  ax.set_title(title)
  ax.set_xlabel(xlabel)
  ax.set_ylabel(ylabel)
  ax.set_zlabel(zlabel)
  plt.show()


plot_1d(f, [-10, 10], 'Plot of f(x) = sin(x)', 'x', 'f(x)')
plot_1d(lambda x: g(x, 0), [-10, 10], 'Plot of g(x, y) = x^4  8x^3 + 24x^2  32x + 16', 'x', 'g(x, y)')
plot_2d(h, [-6, 6], [-6, 6], 'Plot of h(x, y) = exp(-x^2 - y^2)', 'x', 'y', 'h(x, y)')
```
